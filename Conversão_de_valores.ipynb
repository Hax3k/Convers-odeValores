```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

#Cria um controlador de navegador Chrome

navegador = webdriver.Chrome()
navegador.get('https://www.google.com')       # abre navegador

#cotacao dolar 

# Seleciona o campo de pesquisa do google e digita pesquisa
navegador.find_element('xpath', '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input' ).send_keys('Cotação Dolar')


# Aperta enter 
navegador.find_element('xpath', '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input' ).send_keys(Keys.ENTER)

# Obtem valor da moeda 
cotacao_dolar = navegador.find_element('xpath', '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')


#Cotação Euro

navegador.get('https://www.google.com')

navegador.find_element('xpath', '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input' ).send_keys('Cotação Euro')

navegador.find_element('xpath', '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input' ).send_keys(Keys.ENTER)
                                                                                                        
cotacao_euro = navegador.find_element('xpath', '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')


#cotacao do ouro

navegador.get('https://www.melhorcambio.com/ouro-hoje')                                                                                                      
cotacao_ouro = navegador.find_element('xpath', '//*[@id="comercial"]').get_attribute('value')

# Reformata o resultado
cotacao_ouro = cotacao_ouro.replace(',', '.')

#Mostra o resultado da conversão de valores 
print(cotacao_ouro, cotacao_euro, cotacao_dolar)

navegador.quit()                 # Fecha navegador
```

```python
import pandas as pd 

tabela = pd.read_excel('Produtos.xlsx')


display(tabela)       
```

```python
#Localiza as colunas de moedas e atualiza as cotações de acordo com os valores encontrados no google

tabela.loc[tabela['Moeda'] == 'Dólar', 'Cotação'] = float(cotacao_dolar)
tabela.loc[tabela['Moeda'] == 'Euro', 'Cotação'] = float(cotacao_euro)
tabela.loc[tabela['Moeda'] == 'Ouro', 'Cotação'] = float(cotacao_ouro)



tabela['Preço de Compra'] = tabela['Preço Original'] * tabela['Cotação']  #atualiza preço de compra

tabela['Preço de Venda'] = tabela['Preço de Compra'] * tabela['Margem']   #atualiza preço de venda

#Mostra tabela 
display(tabela) 
```

```python
#Cria tabela com valores atualizados 
tabela.to_excel('Produtos Novos.xlsx', index=False)
```
